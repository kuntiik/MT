```python
with open('/home/kuntik/dev/data/dental_rtg_test/annotations/instances_default.json', 'r') as f:
    data = json.load(f)
```

```python
mapping = {}
for i in range(1,101):
    mapping[i] = []
```

```python
for image in data['images']:
    name_s = ''.join(x for x in image['file_name'] if x.isdigit())
    name_int = int(name_s)
    mapping[name_int].append(image['id'])
```
